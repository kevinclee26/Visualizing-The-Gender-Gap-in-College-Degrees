the goal is to generate helpful data visualizations to compare across all degree categories

```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Engineering', 'Computer Science', 'Psychology', 'Biology', 'Physical Sciences', 'Math and Statistics']

fig = plt.figure(figsize=(18, 3))

for sp in range(0,6):
    ax = fig.add_subplot(1,6,sp+1)
    ax.plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    ax.spines["right"].set_visible(False)    
    ax.spines["left"].set_visible(False)
    ax.spines["top"].set_visible(False)    
    ax.spines["bottom"].set_visible(False)
    ax.set_xlim(1968, 2011)
    ax.set_ylim(0,100)
    ax.set_title(stem_cats[sp])
    ax.tick_params(bottom="off", top="off", left="off", right="off")
    
    if sp == 0:
        ax.text(2005, 87, 'Men')
        ax.text(2002, 8, 'Women')
    elif sp == 5:
        ax.text(2005, 62, 'Men')
        ax.text(2001, 35, 'Women')
plt.show()
```

comparing all categories

```python
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']
```

index will be df[0, 1, 2, 3, 4, 5, 6] but location will be sp + 1

pull the last y_value to set the text annotation

```python
max_y_value = women_degrees.loc[len(women_degrees) - 1, 'Business']
print(max_y_value)
```

```python
fig = plt.figure(figsize=(18, 18))

for sp in range(0, 18):
    if not sp == 16: 
        ax = fig.add_subplot(6, 3, sp + 1)
        #select which category array to use
        if (sp % 3) == 0: 
            ax.plot(women_degrees['Year'], women_degrees[stem_cats[int(sp / 3)]], c=cb_dark_blue, label='Women', linewidth=3)
            ax.plot(women_degrees['Year'], 100-women_degrees[stem_cats[int(sp / 3)]], c=cb_orange, label='Men', linewidth=3)
            ax.set_title(stem_cats[int(sp / 3)])
            women_y_value = women_degrees.loc[len(women_degrees) - 1, [stem_cats[int(sp / 3)]]]
        elif (sp % 3) == 1:
            ax.plot(women_degrees['Year'], women_degrees[lib_arts_cats[int(sp / 3)]], c=cb_dark_blue, label='Women', linewidth=3)
            ax.plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[int(sp / 3)]], c=cb_orange, label='Men', linewidth=3)
            ax.set_title(lib_arts_cats[int(sp / 3)])
            women_y_value = women_degrees.loc[len(women_degrees) - 1, [lib_arts_cats[int(sp / 3)]]]
        elif (sp % 3) == 2:
            ax.plot(women_degrees['Year'], women_degrees[other_cats[int(sp / 3)]], c=cb_dark_blue, label='Women', linewidth=3)
            ax.plot(women_degrees['Year'], 100-women_degrees[other_cats[int(sp / 3)]], c=cb_orange, label='Men', linewidth=3)
            ax.set_title(other_cats[int(sp / 3)])
            women_y_value = women_degrees.loc[len(women_degrees) - 1, [other_cats[int(sp / 3)]]]
        #set spines to not visible
        for key,spine in ax.spines.items():
            spine.set_visible(False)
        #set x and y limits
        ax.set_xlim(1968, 2011)
        ax.set_ylim(0,100)
        ax.tick_params(bottom="off", top="off", left="off", right="off")
        #set text annotation
        if float(women_y_value) >= 50: 
            vert_buffer=10
        else: 
            vert_buffer=-10
        if (int(sp / 3)) == 0:
            ax.text(2003, women_y_value + vert_buffer, 'Women')
            ax.text(2005, 100 - women_y_value - vert_buffer, 'Men')
        elif (int(sp / 3)) == 5:
            ax.text(2003, women_y_value + vert_buffer, 'Women')
            ax.text(2005, 100 - women_y_value - vert_buffer, 'Men')
plt.show()
```
